```python
#import

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# where did you get your data from 

data_df=pd.read_csv("SpotifyTopSongsByCountry - May 2020.csv")



```

```python
data_df.head(1000)
```

```python
# How did you clean the data 

clean_df = data_df.dropna(axis= 'rows')
clean_df = clean_df.loc[clean_df['Country'] != 'unknown']
clean_df = clean_df.loc[clean_df['Continent'] != 'unknown']
clean_df = clean_df.loc[clean_df['Rank'] != 'unknown']
clean_df = clean_df.loc[clean_df['Title'] != 'unknown']
clean_df = clean_df.loc[clean_df['Artists'] != 'unknown']
clean_df = clean_df.loc[clean_df['Album'] != 'unknown']
clean_df = clean_df.loc[clean_df['Explicit'] != 'unknown']
clean_df = clean_df.loc[clean_df['Duration'] != 'unknown']


clean_df.count()
```

```python

# Where can I identify the  new clean data 

clean_df.head()
```

```python
from sqlalchemy import create_engine
```

```python
#narrow to display just European data

one_continent = clean_df.loc[clean_df['Continent'] == 'Europe']


one_continent.columns = one_continent.columns.str.lower()
one_continent = one_continent[['country', 'continent', 'rank', 'artists', 'album']]
```

```python
#SQLite upload code 

engine = create_engine("sqlite:///Country.sqlite")
one_continent.to_sql(name="country_Europe", con=engine, if_exists="append", index=False)
pd.read_sql_query("select * from country_Europe", con=engine).head()
```
